```python
import cv2 as cv
```

```python
from PIL import Image
import matplotlib.pyplot as plt
#plot 1:
plt.subplot(2, 2, 1)
im = cv.imread('./sample.jpg')
im = cv.cvtColor(im, cv.COLOR_BGR2RGB)
imgray = cv.cvtColor(im, cv.COLOR_BGR2GRAY)
plt.imshow(im)
plt.title("color")
# plt.xticks([]), plt.yticks([])  #  x  y 

#plot 2:

# 
r = im[:,:,0]
g = im[:,:,1]
b = im[:,:,2] 
br = cv.subtract(b,r)
plt.subplot(2, 2, 2)
plt.imshow(br,cmap = 'gray')
plt.title("gray")
plt.xticks([]), plt.yticks([])  #  x  y 

#plot 3:

plt.subplot(2, 2, 3)
plt.imshow(b,cmap='gray')
plt.title("gray")
plt.xticks([]), plt.yticks([])  #  x  y 


#plot 4:

plt.subplot(2, 2, 4)
plt.imshow(r,cmap='gray')
plt.title("4")
plt.xticks([]), plt.yticks([])  #  x  y 
plt.show()
```
